###$${\textbf{Signals and Plotting the signal.}}$$ 

In Data communication, a signal is an electrical or electromagnetic current that is used for carrying information from one device or network to another.  Periodic analog signals and nonperiodic digital signals are commonly used.

A sine wave shows how the amplitude of a variable changes with time.
$$x(t) = A\sin(2\pi f t + phase)$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def create_sinusoidal(A, f, time, phase = 0):
  x = A * np.sin(2.0 * np.pi * f * time + phase)
  return x
```

```python
time = np.linspace(0, 4.0, 100)
sig = create_sinusoidal(1, 0.5, time)

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))
fig.suptitle('Different subplots of a sinusoidal signal', fontsize=15)

axs[0, 0].plot(time, sig)
axs[0, 0].set_title('Continuous plot', fontsize = 15)
axs[0, 0].set_xlabel('Time', fontsize=12, x = 1)
axs[0, 0].set_ylabel('Amplitude', fontsize=12)

axs[0, 1].stem(time,sig, use_line_collection = "True")
axs[0, 1].set_title('Stem plot', fontsize = 15)
axs[0, 1].set_xlabel('Time', fontsize=12, x = 1)
axs[0, 1].set_ylabel('Amplitude', fontsize=12)

axs[1,0].scatter(time,sig)
axs[1,0].set_title('Scatter plot', fontsize = 15)
axs[-1,-1].axis('off')
axs[1, 0].set_xlabel('Time', fontsize=12, x = 1)
axs[1, 0].set_ylabel('Amplitude', fontsize=12)

plt.show()
```

A single-frequency sine wave is not useful in data communications; 
we need to send a **composite signal**, a signal made of many simple sine waves.

```python
sig1 = create_sinusoidal(1,1,time)
sig2 = create_sinusoidal(2,1.5,time)
sig3 = create_sinusoidal(3,3,time)

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))
fig.suptitle('3 Sinusoidal signal and a composite signal', fontsize=15)

axs[0, 0].plot(time, sig1, 'b')
axs[0, 0].set_title('$sig_1(t)$', fontsize = 15)

axs[0, 1].plot(time, sig2, 'g')
axs[0, 1].set_title('$sig_2(t)$', fontsize = 15)

axs[1,0].plot(time, sig3, 'm')
axs[1,0].set_title('$sig_3(t)$', fontsize = 15)

axs[1,1].plot(time, sig1 + sig2 + sig3)
axs[1,1].set_title('$y(t) = sig_1(t) + sig_2(t) + sig_3(t)$', fontsize = 15)
plt.show()
```
