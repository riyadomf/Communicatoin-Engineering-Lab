###$${\textbf{Pulse Code Modulation.}}$$ 

$$\textrm{ A program to simulate the analog-digital signal using Pulse Code Modulation.}$$

A **PCM encoder** has three processes:  


1.   The analog signal is sampled.
2.   The sampled signal is quantized.
3.   The quantized values are encoded as streams of bits.

We will convert the following analog signal, $x(t) = Asin(2\pi ft +phase)$ into digital signal using Pulse Code Modulation. 






```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''
A - Amplitude of the signal
f - Frequency of the signal
t - Time variable of the signal
T - Time period of the signal
'''
t = np.linspace(0, 4.0, 100)
def create_sinusoidal(A, f, t, phase = 0):
  x = A * np.sin(2.0 * np.pi * f * t + phase)
  return x
```

```python
x_t = create_sinusoidal(1,1,t)
fig, ax = plt.subplots(figsize=(20,5))
ax.plot(t, x_t)
ax.set_title('Analog signal', fontsize = 15)
ax.set_xlabel('Time', fontsize=12)
ax.set_ylabel('Amplitude', fontsize=12)
plt.show()
```

**Process 01: Sampling**  
The analog signal is sampled every $T_s$ second, where $T_s$ is the sample interval or period. The inverse of the sampling interval is called the sampling
rate or sampling frequency and denoted by $f_s$, where $f_s$ = 1/$T_s$.

```python
fs = 20
Ts = 1.0/fs
n = np.arange(0, 4.1, Ts)
x_n = create_sinusoidal(1,1,n)
fig, ax = plt.subplots(figsize=(20,5))
ax.stem(n, x_n, use_line_collection = 'True')
ax.set_title('Sampling Frequency, $f_s=20$', fontsize = 15)
plt.show()
```

**Process 02: Quantization**

The result of sampling is a series of pulses with amplitude values between the maximum
and minimum amplitudes of the signal. The set of amplitudes can be infinite with
nonintegral values between the two limits. These values cannot be used in the encoding
process.  
1. Assuming that the original analog signal has instantaneous amplitudes between
$V_{min}$ and $V_{max}$.  
2. Dividing the range into L zones, each of height Δ (delta).  
$ Δ = \frac{V_{max}-V_{min}}{L}$  
3. Assigning quantized values of 0 to L − 1 to the midpoint of each zone.  
4. Approximating the value of the sample amplitude to the quantized values.  


```python
numberOfBit = 3
L = 2**numberOfBit
V_max=1;
V_min=-V_max;
delta=(V_max-V_min)/L;
partition = np.arange(V_min, V_max, delta)
midpoint= np.arange(V_min-(delta/2),V_max+(delta/2), delta) 

x_q = np.copy(x_n)
encode = np.copy(x_n)

for i in range(len(partition)):
  upper = midpoint[i]
  lower = midpoint[i + 1]
  for j in range(len(x_n)):
    if(i != len(partition) - 1 and upper <= x_n[j] and x_n[j] < lower):
      x_q[j] = partition[i]
      encode[j] = i
    if(i == len(partition) - 1 and upper <= x_n[j]):
      x_q[j] = partition[i]
      encode[j] = i

for i in range(len(x_q)):
    x_q[i] = x_q[i] + (delta / 2) 

fig, ax = plt.subplots(figsize=(20,5))
ax.stem(n, x_q, use_line_collection = 'True')
ax.set_title('Quantized signal', fontsize = 15)
plt.show()
```

**Process 03: Encoding**

The last step in PCM is encoding. After each sample is quantized and the number of bits per sample is decided, each sample can be changed to an $n_b$-bit code word. A quantization code of 2 is
encoded as 010; 5 is encoded as 101; and so on. So the processes for encoding are:   

1.   Convert the assigned decimal codec to the binary codec.
2.   Append all the binary codec into an array of 0 and 1.
3.   Plot a step graph.


```python
def decimalToBinary(value, numberOfBit):
    binary = "{0:b}".format(int(value))
    if(len(binary) > numberOfBit):
      return binary
    extraZero = numberOfBit - len(binary)
    return extraZero * "0" + binary

binary_code = []
digital = []
for i in range(len(encode)):
  binary = decimalToBinary(encode[i], numberOfBit)
  digital.append(binary)
  for j in range(len(binary)):
    binary_code.append(binary[j])
print("Encoded Output: ")
print(digital)
pos = np.arange(len(binary_code))

fig, ax = plt.subplots(figsize=(20,5))
plt.ylim(-1, 2)
ax.step(pos, binary_code)
ax.set_title('Digital signal', fontsize = 15)
plt.show()
```
