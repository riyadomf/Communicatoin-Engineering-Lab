$$\textsf{Phase Modulation}$$

```python
modulator_frequency = 4
carrier_frequency = 40.0
modulation_index = 1.0

time = np.arange(0,1,0.001)
modulator = np.sin(2.0 * np.pi * modulator_frequency * time)
carrier = np.sin(2.0 * np.pi * carrier_frequency * time)
product = np.cos(2*np.pi*carrier_frequency*time + 5*modulation_index*modulator)

fig, axs = plt.subplots(nrows=3, ncols=1, figsize=(10, 8))
fig.suptitle('Phase Modulation', y=1.05, fontsize=18)
plt.setp(axs, yticks=[-1, 0, 1])

axs[0].plot(modulator)
axs[0].set_title('Modulating signal', fontsize = 15)
axs[0].set_xlabel('Time', fontsize=12, x = 1)
axs[0].set_ylabel('Amplitude', fontsize=12)

axs[1].plot(carrier)
axs[1].set_title('Carrier signal', fontsize = 15)
axs[1].set_xlabel('Time', fontsize=12, x = 1)
axs[1].set_ylabel('Amplitude', fontsize=12)

axs[2].plot(product)
axs[2].set_title('Output signal', fontsize = 15)
axs[2].set_xlabel('Time', fontsize=12, x = 1)
axs[2].set_ylabel('Amplitude', fontsize=12)


fig.tight_layout(pad = 2)
plt.show()

```
