###$${\textbf{Basic properties of a signal.}}$$ 

**1) Shifting of a signal**

Time-shifting operation results in the change of just the positioning of the signal without affecting its amplitude or span.

```python
def shifting(y, k, n):
  res=np.zeros(abs(k))

  if(k>=0):
    res=np.concatenate((res,y[0:len(n)-k]))
  else:
    res=np.concatenate((y[abs(k):len(n)],res))
  return res
```

```python
fs = 10
step = 1.0/fs
n = np.arange(0, 2.1, step)
y = create_sinusoidal(1,5/fs,n)
k=6

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

axs[0,0].stem(n, y, use_line_collection = 'True')
axs[0,0].set_title('Original signal, $f_s=10$')
axs[0,0].set_xlabel('Time', fontsize=12, x = 1)
axs[0,0].set_ylabel('Amplitude', fontsize=12)
axs[0,0].set_xticks(np.arange(0, 2.1, step))
axs[0,0].axes.xaxis.set_ticklabels([])



axs[1,0].stem(n,shifting(y,k,n), use_line_collection = 'True')
axs[1,0].set_title('Delay')
axs[1,0].set_xlabel('Time', fontsize=12, x = 1)
axs[1,0].set_ylabel('Amplitude', fontsize=12)
axs[1,0].set_xticks(np.arange(0, 2.1, step))
axs[1,0].axes.xaxis.set_ticklabels([])

k=-6
axs[1,1].stem(n,shifting(y,k,n), use_line_collection = 'True')
axs[1,1].set_title('Advance')
axs[1,1].set_xlabel('Time', fontsize=12, x = 1)
axs[1,1].set_ylabel('Amplitude', fontsize=12)
axs[1,1].set_xticks(np.arange(0, 2.1, step))
axs[1,1].axes.xaxis.set_ticklabels([])

axs[0,1].axis('off')

fig.tight_layout(pad = 2)
plt.show()
```

**2) Folding or Reflection**

If we replace the independent variable n with –n then it is known as folding of signal or it is said to be reflection of the signal about the origin n=0.

```python
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))
axs[0].stem(n,y, use_line_collection = 'True')
axs[0].set_title("Orginal Signal")
axs[0].set_xlabel('Time', fontsize=12, x = 1)
axs[0].set_ylabel('Amplitude', fontsize=12)
axs[0].axes.xaxis.set_ticklabels([])

axs[1].stem(-n,y, use_line_collection = 'True')
axs[1].set_title("Folded Signal")
axs[1].set_xlabel('Time', fontsize=12, x = 1)
axs[1].set_ylabel('Amplitude', fontsize=12)
axs[1].axes.xaxis.set_ticklabels([])

fig.tight_layout(pad = 2)
plt.show()
```
