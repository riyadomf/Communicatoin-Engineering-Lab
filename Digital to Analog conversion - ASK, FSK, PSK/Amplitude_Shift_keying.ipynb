$$\textsf{Amplitude Shift Keying}$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ASK(digital_data):
  #Message_signal
  time = np.arange(0, len(digital_data), 0.001)
  voltage = []
  flag = 0
  for i in digital_data:
    if i==0:
      for j in np.arange(flag, flag+1, 0.001):
        voltage.append(0)
    else:
      for j in np.arange(flag, flag+1, 0.001):
        voltage.append(1)
    flag = flag + 1
  
  #Carrier_signal
  timec = np.linspace(0, 5, len(voltage))
  carrier = np.cos(2*np.pi*10*timec)

  #Modulated_signal
  modulated = voltage*carrier

  #plot
  fig, axs = plt.subplots(nrows=3, ncols=1, figsize=(10, 8))
  fig.suptitle('Amplitude Shift Keying', y=1.05, fontsize=18)
  for ax in axs:
    ax.axes.xaxis.set_ticklabels([])
    ax.axes.yaxis.set_ticklabels([])
    for spine in ['top', 'right']:
      ax.spines[spine].set_visible(False)
    ax.set_xlabel('Time', fontsize=12, x = 1)
    ax.set_ylabel('Amplitude', fontsize=12)

  plt.setp(axs, xticks=np.arange(0,len(digital_data)+1,1))

  axs[0].plot(time, voltage)
  axs[0].set_title('Message signal', fontsize = 15)
  
  axs[1].plot(time, carrier)
  axs[1].set_title('Carrier signal', fontsize = 15)

  axs[2].plot(time, modulated)
  axs[2].set_title('ASK Modulated signal', fontsize = 15)

  fig.tight_layout(pad=1.0)
  plt.show()
```

```python
digital_data = [1, 0, 1, 0, 1, 1, 0, 1]
ASK(digital_data)
```
