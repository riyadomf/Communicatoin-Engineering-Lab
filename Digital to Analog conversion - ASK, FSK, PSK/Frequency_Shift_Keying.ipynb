$$\textsf{Frequency Shift Keying}$$

```python
def FSK(digital_data):
  #Message_signal
  time = np.arange(0, len(digital_data), 0.001)
  voltage = []
  modulated = []
  flag = 0
  for i in digital_data:
    if i==0:
      for j in np.arange(flag, flag+1, 0.001):
        voltage.append(0)
    else:
      for j in np.arange(flag, flag+1, 0.001):
        voltage.append(1)
    flag = flag + 1
  
  #Carrier_signal
  timec = np.linspace(0, 5, len(voltage))
  carrier1 = np.cos(2*np.pi*5*timec)
  carrier2 = np.cos(2*np.pi*20*timec)

  #Modulated_signal
  for i in range(0, len(voltage)):
    if voltage[i]==0:
      temp = np.array([carrier1[i]])
      modulated = np.concatenate((modulated, temp))
    else:
      temp = np.array([carrier2[i]])
      modulated = np.concatenate((modulated, temp))
  
  #plot
  fig, axs = plt.subplots(nrows=4, ncols=1, figsize=(10, 10))
  fig.suptitle('Frequency Shift Keying', y=1.05, fontsize=18)
  for ax in axs:
    ax.axes.xaxis.set_ticklabels([])
    ax.axes.yaxis.set_ticklabels([])
    for spine in ['top', 'right']:
      ax.spines[spine].set_visible(False)
    ax.set_xlabel('Time', fontsize=12, x = 1)
    ax.set_ylabel('Amplitude', fontsize=12)

  plt.setp(axs, xticks=np.arange(0,len(digital_data)+1,1))

  axs[0].plot(time, voltage)
  axs[0].set_title('Message signal', fontsize = 15)
  
  axs[1].plot(time, carrier1)
  axs[1].set_title('1st Carrier signal', fontsize = 15)

  axs[2].plot(time, carrier2)
  axs[2].set_title('2nd Carrier signal', fontsize = 15)

  axs[3].plot(time, modulated)
  axs[3].set_title('FSK Modulated signal', fontsize = 15)

  fig.tight_layout(pad=1.0)
  plt.show()
```

```python
digital_data = [1, 0, 0, 1, 0, 1, 0, 1]
FSK(digital_data)
```
